```python
%%sql
CREATE TABLE IF NOT EXISTS products (
  product_id INTEGER PRIMARY KEY,
  name TEXT,
  price REAL
)
STRICT;
```

```python
%%sql
INSERT INTO products (name, price) VALUES ('Superbowl Jersey', 120.00);
INSERT INTO products (name, price) VALUES ('Team Cap', 24.99);
INSERT INTO products (name, price) VALUES ('Hoodie', 59.50);
SELECT * FROM products ORDER BY product_id;
```

```python
%%sql
SELECT * FROM products
WHERE price > 50
ORDER BY price DESC;
```

```python
%%sql
UPDATE products
SET price = 99.99
WHERE name = 'Superbowl Jersey';
SELECT * FROM products ORDER BY product_id;
```

```python
%%sql
DELETE FROM products
WHERE product_id = 1;
SELECT * FROM products ORDER BY product_id
```
